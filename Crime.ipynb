```python
#import the libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import the data

df = pd.read_csv('Crime_Data.csv')
df.head(40)
```

```python
#rename column
data = df.rename(columns={'Date Rptd':'DateRptd'})
data.head(10)
```

```python
#split the DateRptd clumn into date and time
Dtime = data.DateRptd.str.split(expand=True)
Dtime.head(20)
```

```python
data['DateRptd'] = Dtime[0].values
data.head()
```

```python
Docc = data["DATE OCC"].str.split(expand=True)
Docc.head()
```

```python
data["DATE OCC"] = Docc[0].values
data.head()
```

```python
dfdrop = data
```

```python
#axis =1 means drop columns
DATA = dfdrop.drop(['Mocodes', 'Crm Cd 1' , 'Crm Cd 2' , 'Crm Cd 3' , 'Crm Cd 4', 'Cross Street'],axis=1)
```

```python
#to see the remaining columns
DATA.columns
```

```python
#no of null values in each column is equal to..
DATA.isnull().sum()
```

```python
DATAS = DATA.fillna('NAN')
```

```python
DATAS.isnull().sum()
```

```python
DATAS.head(10)
```

```python
crime_type_counts = DATAS['Crm Cd Desc'].value_counts()
top_10_crime_types = crime_type_counts.head(10)
plt.figure(figsize=(8, 8))
plt.pie(top_10_crime_types, labels=top_10_crime_types.index, autopct='%1.1f%%')
plt.title('Top 10 Crime Types Distribution')
plt.savefig('piechart.png', dpi=500)
plt.show()
```
